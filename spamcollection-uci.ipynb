Data Clearning and Data Preprocessing
In messages we have questions mark colon and comma and not required to classify 
as spam or not

```python
import re
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
corpus=[]
for i in range(0,len(df_Spamdata)):
    review = re.sub('[^a-zA-Z]', ' ', df_Spamdata['message'][i])
    review=review.lower()
    review=review.split()
    review=[ ps.stem(word) for word in review if not word in stopwords.words('english')]
    
    review= ' '.join(review)
    corpus.append(review)

#PortStemmer used for stemming purpose. Stemming helps to find base root of word
```

```python
#Create bag of words to just see in the form of importance of 0 and 1
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000)
X=cv.fit_transform(corpus).toarray()
```

```python
pd.DataFrame(X)
```

```python
y=pd.get_dummies(df_Spamdata['label'])
y
```

```python
y=y.iloc[:,1].values
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
from sklearn.naive_bayes import MultinomialNB
spam_detect_model = MultinomialNB().fit(X_train, y_train)
```

```python

y_pred=spam_detect_model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
confusion_m=confusion_matrix(y_test,y_pred)
```

```python
confusion_m
```

```python
from sklearn.metrics import accuracy_score
accuracy_Score=accuracy_score(y_test,y_pred)
```

```python
accuracy_Score
```
